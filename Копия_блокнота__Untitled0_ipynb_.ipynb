```python
# --- Импорт библиотек ---
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import requests
from datetime import datetime, timedelta
import numpy as np
```

```python
# --- Настройка графиков ---
plt.style.use('default') # Или 'seaborn-whitegrid'
sns.set_palette("husl")
```

```python
# --- Получение данных ---
# Используем API Центробанка РФ за последние N дней (например, 30)
# API может быть ограничено, поэтому часто используются сторонние сервисы или исторические данные.
# ВНИМАНИЕ: Этот пример использует XML API ЦБ РФ, который требует немного больше обработки.
# Более простой способ - найти CSV файл с историческими данными или использовать другой API.
```

```python
# Пример получения данных за один день (XML) - сложнее для парсинга
# url_today = f"https://www.cbr.ru/scripts/XML_daily.asp?date_req={datetime.today().strftime('%d/%m/%Y')}"
```

```python
# Более простой путь - получить данные за диапазон в JSON (если доступно, или CSV)
# Многие ресурсы предоставляют исторические данные.
# Давайте смоделируем небольшой датафрейм с искусственными, но реалистичными данными курса USD/RUB за последние 30 дней.
print("Генерация искусственных данных курса USD/RUB за последние 30 дней...")

start_date = datetime.today() - timedelta(days=30)
dates = pd.date_range(start=start_date, end=datetime.today(), freq='D')
```

```python
# Генерируем случайные колебания вокруг базового значения (например, 90 руб.)
base_rate = 90.0
volatility = 0.5 # уровень волатильности
rates = base_rate + np.cumsum(np.random.normal(0, volatility, size=len(dates))) # случайное блуждание

df_currency = pd.DataFrame({
    'Date': dates,
    'USD_RUB': rates
})
```

```python
# Конвертируем Date в datetime, если еще не
df_currency['Date'] = pd.to_datetime(df_currency['Date'])
```

```python
# --- Исследовательский анализ данных (EDA) ---
print("\n--- Общая информация о датасете ---")
print(df_currency.info())

print("\n--- Первые 5 строк ---")
print(df_currency.head())

print("\n--- Последние 5 строк ---")
print(df_currency.tail())

print("\n--- Описательная статистика ---")
print(df_currency.describe())

print("\n--- Проверка на пропущенные значения ---")
print(df_currency.isnull().sum())
```

```python
# --- Визуализация ---
# Настройка размера графика
plt.figure(figsize=(12, 6))
```

```python
# Линейный график курса USD/RUB
plt.plot(df_currency['Date'], df_currency['USD_RUB'], marker='', linestyle='-', color='b', label='USD/RUB')
plt.title('Курс доллара США (USD) к рублю (RUB) за последние 30 дней')
plt.xlabel('Дата')
plt.ylabel('Курс (RUB)')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45) # Поворот подписей оси X для лучшей читаемости
plt.tight_layout() # Подгонка элементов графика
plt.show()
```

```python
# --- Дополнительно (пример) ---
# Расчет ежедневного изменения в процентах
df_currency['Daily_Change_Pct'] = df_currency['USD_RUB'].pct_change() * 100

plt.figure(figsize=(12, 4))
plt.bar(df_currency['Date'], df_currency['Daily_Change_Pct'], color='orange', alpha=0.7)
plt.title('Ежедневное изменение курса USD/RUB (%)')
plt.xlabel('Дата')
plt.ylabel('Изменение (%)')
plt.grid(axis='y')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print("\n--- Среднее ежедневное изменение ---")
avg_change = df_currency['Daily_Change_Pct'].abs().mean()
print(f"Средняя абсолютная величина дневного изменения: {avg_change:.4f}%")
```

```python
# --- Сохранение ---
# Сохраняем полученный DataFrame в файл CSV
output_filename = 'currency_rates_usd_rub_30days.csv'
df_currency.to_csv(output_filename, index=False)
print(f"\nДанные сохранены в файл: {output_filename}")
```
